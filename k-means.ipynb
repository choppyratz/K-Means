```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.spatial import distance
%matplotlib inline
plt.rcParams["figure.figsize"] = (7, 5)
```

```python
x = np.random.randint(1,7000,20000) #x coordinates of objects
y = np.random.randint(1,7000,20000) #y coordinates of objects
x_c = np.random.randint(1,7000,20) #x coordinates of classes
y_c = np.random.randint(1,7000,20) #y coordinates of classes
generations = 5
classes = []
colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#e6beff', '#9a6324', '#fff799', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080']

class clust:
    def __init__(self, color, x, y):
        self.color = color
        self.x_cords = []
        self.y_cords = []
        self.x = x
        self.y = y
        
```

```python
plt.scatter(x, y, s=2, c='#000000')
for i in range(20):
    color = colors[np.random.randint(0, 19,1)[0]]
    classes.append(clust(color, x_c[i], y_c[i]))
    plt.scatter(classes[i].x, classes[i].y, s=80, c=classes[i].color, marker='s')
plt.show()
```

```python
def det_class_for_points(x_arr, y_arr, classes_arr):
    for ind, i in enumerate(x_arr):
        min_dist = -1
        min_dist_class = None
        for j in classes_arr:
            dis = distance.euclidean([x_arr[ind], y_arr[ind]], [j.x, j.y])
            if ((dis < min_dist) or (min_dist == -1)):
                min_dist = dis
                min_dist_class = j
        min_dist_class.x_cords.append(x_arr[ind])
        min_dist_class.y_cords.append(y_arr[ind])
```

```python
def visualize_clusters(classes_arr):
    for i in classes_arr:
        plt.scatter(i.x_cords, i.y_cords, s=2, c=i.color)
        plt.scatter(i.x, i.y, s=30, c='#000000', marker='s')
    plt.show()
```

```python
def override_clusters_centrs(classes_arr):
    for i in classes_arr:
        i.x = sum(i.x_cords) / len(i.x_cords)
        i.y = sum(i.y_cords) / len(i.y_cords)
```

```python
def clear_clusters_points(classes_arr):
    for i in classes_arr:
        i.x_cords = []
        i.y_cords = []
```

```python
for i in range(1, generations):
    clear_clusters_points(classes)
    det_class_for_points(x, y, classes)
    override_clusters_centrs(classes)
    visualize_clusters(classes)
```
